Manejo de Strings en python

String es una composicion de caracteres, que nos permite hacer indexaciones. Muy parecido a la tupla

```python
String_1 ="Hola como estas"
```

```python
String_1[3] #Esta es la posicion 3 de la cadena 
```

```python
String_1[1:4] #Esto es desde la posicion 1 hasta la 4 (sin incluir la 4)
String_1[4:] #Desde la posicion 4 hasta el final
String_1[:4] #Desde el inicio hasta la posicion 4 (sin incluir la 4)
String_1[1: :2] #Desde la posicion 1 hasta el final, de 2 en 2
```

El string es inmutable

```python
String_1
```

```python
String_1.upper() #Convierte todo a mayusculas
String_1.lower() #Convierte todo a minusculas
String_1.replace("o","a") #Reemplaza las "o" por "a"
String_1.split(" ") #Separa la cadena en una lista, usando el espacio como separador
String_1.find("como") #Busca la palabra "como" y devuelve la posicion inicial
```

```python
String_2 = String_1.replace(" ","")
print(String_2)
```
